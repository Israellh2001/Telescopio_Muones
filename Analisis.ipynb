```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
np.set_printoptions(suppress=True)
```

```python
datos = np.genfromtxt("/home/izheynes/Code/telescopioMuones/build/Histograma_nt_Hits.csv", delimiter=",")
datosProcess = np.unique(datos, axis=0)

sensors = np.array([11,12,13,21,22,23])
puntos = []
stripY = []
stripX = []
strips = []

for dato in datosProcess:
    if ((dato[2] < 19999) or ((dato[2] > 29999) and (dato[2]<39999))):
        stripX.append(dato)
    else:
        stripY.append(dato)
    strips.append(dato)
    
stripY = np.array(stripY)
stripX = np.array(stripX)
strips = np.array(strips)

for sensor in sensors:
    for Y in stripY[stripY[:,1]==sensor]:
        for X in stripX[stripX[:,1]==sensor]:
            puntos.append([Y[3],X[4], (Y[5]+X[5])/2])
puntos = np.array(puntos)
np.savetxt("s3.csv", strips, delimiter=",", fmt="%i")
```

```python
datos = np.genfromtxt("/home/izheynes/Code/telescopioMuones/build/Histograma_nt_Hits.csv", delimiter=",")

energia = []

for dato in datos:
    energia.append(dato[6])
energia = np.array(energia)
np.savetxt("Es1.csv", energia, delimiter=",", fmt="%f")
```

```python
%matplotlib widget
fig = plt.figure(figsize = (14,7))
#fig = plt.figure()
ax = plt.axes(projection="3d")

ax.set_ylabel("Y")
ax.set_xlabel("X")
ax.set_zlabel("Z")
ax.scatter3D(puntos[:,0],puntos[:,1],puntos[:,2])

plt.show()
```
